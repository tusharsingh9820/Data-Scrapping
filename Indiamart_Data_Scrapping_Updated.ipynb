# **1.Home Scrapping**

## Extract upper data

```python
def get_company_details(url):
    response = requests.get(url)
    soup1 = BeautifulSoup(response.content, 'html.parser')


    name = soup1.find_all('h1',class_ = 'FM_C0 FM_lh1 FM_Lsp1 FM_f28 FM_bo FM_cpdef')
    name =  [title.text.strip() for title in name]if name else 'N/A'


    trustseal_present = soup1.find_all('p',class_ = 'FM_trs FM_ds15 FM_ds5 FM_Lsp FM_cp') is not None


    gst = soup1.find('span',class_ = 'FM_bo')
    gst =  [title.text.strip() for title in gst]if gst else 'N/A'


    addre = soup1.find('span',class_ = 'FM_p32')
    addre =  [title.text.strip() for title in addre]if addre else 'N/A'


    pro = soup1.find_all('p',class_ = 'fasC FM_w1')
    pro =  [title.text.strip() for title in pro]if pro else 'N/A'


    return {
        'Name': name,
        'Verified': trustseal_present,
        'GST Number': gst,
        'Address': addre,
        'Products': pro,

    }
```

```python
data1 = []
for url in urls:
    details = get_company_details(url)
    data1.append(details)
home = pd.DataFrame(data1)
home
```

##Extract table

```python
def get_table_details(urls):
    response = requests.get(urls)
    soup = BeautifulSoup(response.content, 'html.parser')

    table = soup.find_all('div', class_='FM_m9 FM_m18 FM_AaRt')

    headers = []
    data = []

    for entry in table:
        header = entry.find('p', class_='FM_c6 FM_f15 FM_m13')
        value = entry.find('span', class_='Fm_lh18 FM_f16 FM_c1')

        if header and value:
            headers.append(header.text.strip())
            data.append(value.text.strip())


    detail = pd.DataFrame([data], columns=headers)
    detail['url'] = urls

    return detail
```

```python
df2 = pd.DataFrame()
for url in urls:
    company_detail = get_table_details(url)
    df2 = pd.concat([df2, company_detail], ignore_index=True)
df2
```

## Merging all dataframe into one

```python
home['GST Number'] = home['GST Number'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
df2['GST Number'] = df2['GST Number'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
Home_detail_table = pd.merge(home, df2, on='GST Number', how='inner')
Home_detail_table
```

## Create seperate product cell for each company

```python
Home_detail_table['Product Count'] = Home_detail_table['Products'].apply(lambda x: len(str(x).split(',')))
max_products = Home_detail_table['Product Count'].max()
product_columns = [f'Product {i+1}' for i in range(max_products)]
Home_detail_table[product_columns] = Home_detail_table['Products'].apply(lambda x: pd.Series(str(x).split(',')))
detail_table = Home_detail_table.drop(columns=['Products'])
```

#**2.About Us Scrapping**

##Extracting About us Section

```python
def get_aboutus_link(url):
    response = requests.get(url)
    soup2 = BeautifulSoup(response.content, 'html.parser')

    gst = soup2.find('span',class_ = 'FM_bo')
    gst =  [title.text.strip() for title in gst]if gst else 'N/A'

    patterns = [
        r"^https:\/\/www\.indiamart\.com\/[^\/]+\/aboutus\.html$",
        r"^https:\/\/www\.indiamart\.com\/[^\/]+\/?#\w*$",
        r"^https:\/\/www\.indiamart\.com\/[^\/]+\/profile\.html$",
        r"^https:\/\/www\.indiamart\.com\/[^\/]+\/about-us\.html$",
        r"^https:\/\/www\.indiamart\.com\/company\/\d+\/aboutus\.html$",
        r"^https:\/\/www\.indiamart\.com\/company\/\d+\/about-us\.html$",
        r"^https:\/\/www\.indiamart\.com\/company\/\d+\/profile\.html$",
        r"^https:\/\/www\.indiamart\.com\/company\/\d+\/?#\w*$",
        r"https:\/\/www\.indiamart\.com\/([^\/]+)\/(about-us\.html|aboutus\.html|profile\.html|#aboutus)",
        r"https:\/\/www\.indiamart\.com\/company\/\d+\/(about-us\.html|aboutus\.html|profile\.html|#aboutus)"
    ]

    aboutus_link_tags = soup2.find_all('ul', class_='FM_ds5 FM_f16 FM_w1')
    aboutus_links_list = []

    for tag in aboutus_link_tags:
        about_links = tag.find_all('a', href=True)
        for a in about_links:
            full_url = urljoin(url, a['href'])
            if any(re.match(pattern, full_url) for pattern in patterns):
                aboutus_links_list.append(full_url)


    if not aboutus_links_list:
        return pd.DataFrame()


    full_url = aboutus_links_list[0]
    response = requests.get(full_url)
    soup3 = BeautifulSoup(response.content, 'html.parser')


    table = soup3.find('section', class_='FM_m11 FM_ds20 p010')
    if not table:
        return pd.DataFrame()


    headers = []
    data = []


    for row in table.find_all('tr'):
        cells = row.find_all('td')
        if len(cells) == 2:
            header = cells[0].get_text(strip=True)
            data_field = cells[1]


            ul = data_field.find('ul')
            if ul:

                data_field = ', '.join(li.get_text(strip=True) for li in ul.find_all('li'))
            else:
                data_field = data_field.get_text(strip=True)

            headers.append(header)
            data.append(data_field)


    data_dict = pd.Series(data, index=headers).to_dict()
    about_us_df = pd.DataFrame([data_dict])
    about_us_df['GST Number'] = gst

    return about_us_df
```

```python
About_us_detail = pd.DataFrame()
About_Us= pd.DataFrame()
for url in urls:
    About_us_detail = get_aboutus_link(url)
    About_Us = pd.concat([About_Us, About_us_detail], ignore_index=True)
About_Us
```

#**3.Contact Us Scrapping**

## Getting contact us link into lists

```python
def get_contactus_link(url):
    response = requests.get(url)
    soup3 = BeautifulSoup(response.content, 'html.parser')

    pattern1 = r"^https:\/\/www\.indiamart\.com\/[^\/]+\/enquiry\.html$"
    pattern2 = r"^https:\/\/www\.indiamart\.com\/[a-zA-Z0-9\-]+(?:-[a-zA-Z0-9\-]+)*\/enquiry\.html$"
    pattern3 = r"^https:\/\/www\.indiamart\.com\/(?:[a-zA-Z0-9\-]+\/)?enquiry\.html$"
    pattern4 = r"^https:\/\/www\.indiamart\.com\/[a-zA-Z0-9\-]+\/enquiry\.html$"
    pattern5 = r"https:\/\/www\.indiamart\.com\/([^\/]+)\/enquiry\.html"

    contactus_tags = soup3.find_all('ul', class_='FM_ds5 FM_f16 FM_w1')
    contactus_links_list = []

    for tag in contactus_tags:
        links = tag.find_all('a', href=True)
        for a in links:
            full_url = urljoin(url, a['href'])

            if re.match(pattern1, full_url) or re.match(pattern2, full_url) or re.match(pattern3, full_url) or re.match(pattern4, full_url) or re.match(pattern5, full_url) :
                contactus_links_list.append(full_url)

    return contactus_links_list


data3 = []
for url in urls:
    detail3 = get_contactus_link(url)
    data3.append(detail3)
contactus_links = data3
contactus_links
```

##Extracting Contact us Section

```python
def get_contactus_details(url):
    response = requests.get(url)
    soup4 = BeautifulSoup(response.content, 'html.parser')

    span = soup4.find('span', {'id': 'footerPNS'})

    gst_span = soup4.find('span', class_='FM_bo')
    gst = gst_span.text.strip() if gst_span else 'N/A'


    person_p = soup4.find('p', class_='FM_Lsp4 FM_f15 FM_c7 FM_p29')
    person = person_p.text.strip() if person_p else 'N/A'


    phone_number = 'N/A'
    phone_span = soup4.find('span', class_='FM_c2 FM_tduHvr')
    phone_text = phone_span.get_text(strip=True) if phone_span else 'N/A'
    if span and 'data-pnsno' in span.attrs:
        phone_text = span['data-pnsno']
        phone_number_match = re.search(r'\d{10}', phone_text)
        phone_number = phone_number_match.group() if phone_number_match else 'N/A'

    return {
        'GST Number': gst,
        'Contactable Person': person,
        'Phone Number': phone_number
    }


data4 = []
for sublist in contactus_links:
    for url in sublist:
        details4 = get_contactus_details(url)
        data4.append(details4)
contact = pd.DataFrame(data4)
contact
```
