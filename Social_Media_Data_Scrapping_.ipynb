```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from google.colab import files
from urllib.parse import urljoin
import re
```

#**1. Linked IN**

## i. Create list of LinkedIN links

```python
with open('LinkedIN_urls.txt', 'r') as file:
    content1 = file.read()
LinkedIN_urls = content1.split(',')
LinkedIN_urls = [item.strip() for item in LinkedIN_urls]
LinkedIN_urls
```

##ii. Extract Data

```python
def get_linkedIN_table_details(urls):
    response = requests.get(urls)
    soup = BeautifulSoup(response.content, 'html.parser')

    name = soup.find_all('h1', class_='text-heading-xlarge inline t-24 v-align-middle break-words')
    name = [title.text.strip() for title in name] if name else ['N/A']

    summary = soup.find_all('p', class_='break-words white-space-pre-wrap t-black--light text-body-medium')
    summary = [title.text.strip() for title in summary] if summary else ['N/A']

    table = soup.find_all('div', class_='overflow-hidden')

    headers = []
    data = []

    for entry in table:
        header = entry.find('h3', class_='text-heading-medium')
        value = entry.find('dd', class_='mb4 t-black--light text-body-medium')

        if header and value:
            headers.append(header.text.strip())
            data.append(value.text.strip())


    detail = pd.DataFrame([data], columns=headers)
    detail['url'] = urls
    detail['name'] = name
    detail['summary'] = summary

    return detail

```

```python
LinkedIN = pd.DataFrame()
for url in LinkedIN_urls:
    try:
        LinkedIN_detail_1 = get_linkedIN_table_details(url)
        LinkedIN = pd.concat([LinkedIN, LinkedIN_detail_1], ignore_index=True)
    except Exception as e:
        print(f"Error processing {url}: {e}")

print(LinkedIN)
```

```python
file_path1 = '/content/LinkedIN_data.xlsx'
LinkedIN.to_excel(file_path1, index=False)

files.download(file_path1)
```

#**2. Youtube**

## i. Create list of LinkedIN links

```python
with open('Youtube_urls.txt', 'r') as file:
    content2 = file.read()
Youtube_urls = content2.split(',')
Youtube_urls = [item.strip() for item in Youtube_urls]
Youtube_urls
```

##ii. Extract Data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_youtube_channel_details(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')


    table = soup.find_all('yt-content-metadata-view-model', class_='page-header-view-model-wiz__page-header-content-metadata yt-content-metadata-view-model-wiz yt-content-metadata-view-model-wiz--inline')

    headers = []
    data = []

    for entry in table:

        value = entry.find('span', class_='yt-core-attributed-string yt-content-metadata-view-model-wiz__metadata-text yt-core-attributed-string--white-space-pre-wrap yt-core-attributed-string--link-inherit-color',dir_="auto")

        if  value:

            data.append(value.text.strip())


    detail = pd.DataFrame([data], columns=headers)
    # Create a DataFrame from Youtube_urls to ensure proper alignment
    url_df = pd.DataFrame({'url': Youtube_urls})
    # Concatenate the url DataFrame with the details DataFrame
    detail = pd.concat([detail, url_df], axis=1)

    return detail
```

```python
data2 = []
for url in Youtube_urls:
    Youtube_details = get_youtube_channel_details(url)
    data2.append(Youtube_details)
    Youtube = pd.concat(data2, ignore_index=True)
Youtube
```

```python
file_path2 = '/content/Youtube_data.xlsx'
Youtube.to_excel(file_path1=2, index=False)

files.download(file_path2)
```

#**3. Instagram**

## i. Create list of Instagram links

```python
with open('Instagram_urls.txt', 'r') as file:
    content3 = file.read()
Instagram_urls = content3.split(',')
Instagram_urls = [item.strip() for item in Instagram_urls]
Instagram_urls
```

## i. Create list of Instagram links

```python
def get_instagram_details(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')


    name_element = soup.find('meta', property='og:title')
    name = name_element['content'].strip() if name_element else 'N/A'


    meta_elements = soup.find_all('meta', {'property': 'og:description'})
    followers = 'N/A'
    posts = 'N/A'

    for meta in meta_elements:
        content = meta.get('content', '')
        if 'Followers' in content:
            parts = content.split(' - ')
            if len(parts) > 1:
                data_parts = parts[0].split(', ')
                for part in data_parts:
                    if 'Followers' in part:
                        followers = part.split(' ')[0]
                    elif 'Posts' in part:
                        posts = part.split(' ')[0]

    data = {
        'Company Name': [name],
        'Followers': [followers],
        'Posts': [posts]
    }

    df = pd.DataFrame(data)
    return df
```

```python
Instagram = pd.DataFrame()
for url in Instagram_urls:
    try:
        Instagram_detail_1 = get_instagram_details(url)
        Instagram = pd.concat([Instagram, Instagram_detail_1], ignore_index=True)
    except Exception as e:
        print(f"Error processing {url}: {e}")

Instagram
```

```python
file_path3 = '/content/Instagram_data.xlsx'
Instagram.to_excel(file_path3, index=False)

files.download(file_path3)
```
