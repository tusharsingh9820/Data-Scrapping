```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.indiamart.com/eurobond-industriesmumbai/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup)
```

#Data Scrapping

##Extract name

```python
name = soup.find_all('h1',class_ = 'FM_C0 FM_lh1 FM_Lsp1 FM_f28 FM_bo FM_cpdef')
name =  [title.text.strip() for title in name]
n = ['Name of Company']
Name = pd.DataFrame([name], columns=n)
Name
```

##Scrap Verification trust seal

```python
trustseal_present = soup.find_all('p',class_ = 'FM_trs FM_ds15 FM_ds5 FM_Lsp FM_cp') is not None
trustseal_present
data = {
    'TrustSEAL Verification': ['Yes' if trustseal_present else 'No']
}
verify = pd.DataFrame(data)
verify
```

##Scrapping GST number

```python
gst = soup.find('span',class_ = 'FM_bo')
gst_number =  [title.text.strip() for title in gst]if gst else 'N/A'
G = ['GST number']
GST_number = pd.DataFrame([gst_number], columns=G)
GST_number
```

##Scrapping Address

```python
addre = soup.find('span',class_ = 'FM_p32')
address =  [title.text.strip() for title in addre]
ad = ['Address']
address = pd.DataFrame([address], columns=ad)
address
```

## Scrapping Products

```python
pro = soup.find_all('p',class_ = 'fasC FM_w1')
product =  [title.text.strip() for title in pro]
products_string = ', '.join(product)
products_list = [products_string]
p = ['Products']
Product = pd.DataFrame([products_list], columns=p)
Product
```

##Detail Scrapping

```python
table = soup.find_all('div',class_ = 'FM_m9 FM_m18 FM_AaRt')
headers = []
data = []
for entry in table:
    header = entry.find('p', class_='FM_c6 FM_f15 FM_m13').text.strip()
    value = entry.find('span', class_='Fm_lh18 FM_f16 FM_c1').text.strip()
    headers.append(header)
    data.append(value)
detail = pd.DataFrame([data], columns=headers)
detail
```

## Merging all dataframe into one

```python
df = pd.merge(Name, verify, left_index=True, right_index=True)
df = pd.merge(df, GST_number, left_index=True, right_index=True)
df = pd.merge(df, address, left_index=True, right_index=True)
df = pd.merge(df, Product, left_index=True, right_index=True)
df = pd.merge(df, detail, left_index=True, right_index=True)
```
